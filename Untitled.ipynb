Now I will perform a hypothesis test at 5% significace.
I will use a t test to compare the average scores between the top threshold of students and the average of the students
I will sample my students from the top 5% of scores

H0: More study hours do not lead to higher scores
H1: Students who study the most score higher than average

```python
top_threshold = data['Hours_Studied'].quantile(0.95)
top_students = data[data['Hours_Studied'] >= top_threshold]

from scipy.stats import ttest_ind
rest_students = data[data['Hours_Studied'] < top_threshold]

t_stat, p_value = ttest_ind(top_students['Exam_Score'], rest_students['Exam_Score'], equal_var=False)
print('t-statistic:', t_stat)
print('p-value:', p_value)
```

As we can see from the t-statistic being a high positive value, the top students' scores are much higher than the rest.
The p value is very low which makes it extremely unlikely that this difference is random. Strong evidence to suggest that studying more is associated with higher scores.
